```python
import matplotlib.pyplot as plt
import numpy as np
# xpoinst = [-3,-2,-1,0,1,2,3]
#xpoinst = np.arange(-5,2,0.5)
xpoinst = np.linspace(-5,2,0.5)
ypoints = [-x**2 + 3 * x+7 for x in xpoinst]

plt.plot(xpoinst, ypoints, c='r', marker = 'o', linestyle = '--', lw = 2, label="-x**2 + 3 * x+7")
plt.legend()
plt.title("Wykres funkcji kwadratowej")
plt.xlabel("Oś X")
plt.ylabel("Oś Y")
plt.grid()


plt.show()
```

```python
rolls_count = []
dice_value = []
dice_rolls = np.array([(np.random.randint(1,7)) for _ in range(20)])
print(dice_rolls)

for i in set(dice_rolls):
  dice_value.append(i)
  rolls_count.append(len(np.where(dice_rolls == i)[0]))
plt.bar(dice_value, rolls_count)
plt.show()
```
